```python
def f(x): #example 2
    x = x + 1
    print(x)

n = 5
f(n) #x changed
print(n) #n did not change
```

```python
#example 2 questions

#was the object mutable or immutable?
#n is immutable, x is mutable

#Did Python mutate the object or rebind the name?
#nothing was done to n, x was mutated
```

```python
x = 5 #example 3
print(id(x))
x += 1
print(id(x))
```

```python
#example 3 questions

#was the object mutable or immutable?
#x is immutable

#Did Python mutate the object or rebind the name?
#x was rebound to a new object
```

```python
lst = [1, 2] #example 4
print(id(lst))
lst.append(3)
print(id(lst)) #same list; list is mutable
```

```python
#example 4 questions

#Was the object mutable or immutable?
#mutable

#Did Python mutate the object or rebind the name?
#the list was mutate, as it has the same ID
```

```python
def change(x): #example 5
    x = x + 10

a = 5
change(a)
print(a)
```

```python
#example 5 questions

#Was the object mutable or immutable?
#a is immutable

#Did Python mutate the object or rebind the name?
#a never got changed, but x is mutated.
```

```python
def change(lst): #example 6
    lst.append(99)

data = [1,2]
change(data)
print(data)
```

```python
#example 6 questions

#Was the object mutable or immutable?
#the object is mutable

#Did Python mutate the object or rebind the name?
#it was mutated
```

```python
def reset_wrong(lst): #example 7
    lst = []

def reset_right(lst):
    lst.clear()

data = [1,2,3]
reset_wrong(data)
print(data)

reset_right(data)
print(data)
```

```python
#example 7 questions

#Was the object mutable or immutable?
#the global lst is mutable, the local list in the wrong function is also mutable

#Did Python mutate the object or rebind the name?
#python mutated the global lst
```
